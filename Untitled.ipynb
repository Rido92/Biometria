```python
#limpeza dos do dataset: apageui as colunas massa estimadas e comprimento. substitui os valores ausentes da coluna semana de cultivo
data=data.drop(['comprimento', 'Massa estimada'],axis=1)
```

```python
import cv2
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=data.Biometria
x=data.drop('Biometria', axis=1)
```

```python

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
xtest
```

```python
xtrain
```

```python
# Preprocessing – Scaling the features
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

x_train_scaled = scaler.fit_transform(xtrain)
x_train = pd.DataFrame(x_train_scaled)

x_test_scaled = scaler.fit_transform(xtest)
x_test = pd.DataFrame(x_test_scaled)
y_train=ytrain
y_test=ytest
```

```python
from sklearn.linear_model import LinearRegression as lm
model=lm().fit(x_train,y_train)
model.score(x_train,y_train)

```

```python
from sklearn.neighbors import KNeighborsRegressor
model=KNeighborsRegressor(n_neighbors=2)
model=model.fit(x_train,y_train)
model.score(x_train,y_train)
predictions=model.predict(x_test)
import matplotlib.pyplot as plt
plt.scatter(y_test,predictions, c='r')
plt.ylabel('Predictions')
plt.xlabel('trues values')
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor
model=SGDRegressor(max_iter=1000, tol=1e-5)
model=model.fit(x_train,y_train)
model.score(x_train,y_train)
predictions=model.predict(x_test)
import matplotlib.pyplot as plt
plt.scatter(y_test,predictions, c='r')
plt.ylabel('Predictions')
plt.xlabel('trues values')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(max_depth=2, random_state=0)
model=model.fit(x_train,y_train)
model.score(x_train,y_train)
predictions=model.predict(x_test)
import matplotlib.pyplot as plt
plt.scatter(y_test,predictions, c='r')
plt.ylabel('Predictions')
plt.xlabel('trues values')
plt.show()
```

```python
from sklearn.svm import SVR
model=SVR(C=10000.0, epsilon=0.2)
model=model.fit(x_train,y_train)
model.score(x_train,y_train)
predictions=model.predict(x_test)
import matplotlib.pyplot as plt
plt.scatter(y_test,predictions, c='r')
plt.ylabel('Predictions')
plt.xlabel('trues values')
plt.show()
```
